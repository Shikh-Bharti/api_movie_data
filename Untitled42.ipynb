```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Extraction
def fetch_movie_data(api_key, movie_title):
    url = f"http://www.omdbapi.com/?t={movie_title}&apikey={api_key}"
    response = requests.get(url)
    return response.json()

# List of movies to fetch data for
movies = ["Inception", "The Matrix", "Interstellar", "The Godfather", "Pulp Fiction"]
movie_data = []

# Fetch data for each movie
for movie in movies:
    data = fetch_movie_data("api_key", movie)  # Replace YOUR_API_KEY with your actual API key
    if 'Error' not in data:  # Check for errors in the response
        movie_data.append(data)
    else:
        print(f"Error fetching data for {movie}: {data['Error']}")

# Step 2: Data Cleaning and Manipulation
df_movies = pd.DataFrame(movie_data)

# Print the columns to understand the structure
print("Columns in the DataFrame:", df_movies.columns)

# Adjust the columns based on the actual keys returned by the API
# For example, if the response has 'Genre' as 'Genre', use it directly.
df_movies = df_movies[['Title', 'Year', 'Genre', 'imdbRating']]  # Adjust as necessary
df_movies['imdbRating'] = pd.to_numeric(df_movies['imdbRating'], errors='coerce')  # Convert rating to numeric
df_movies.dropna(inplace=True)  # Drop rows with missing values

# Step 3: Data Analysis
# Calculate average ratings for different genres
df_movies['Genres'] = df_movies['Genre'].str.split(", ")  # Split genres into lists
genre_exploded = df_movies.explode('Genres')  # Explode genres into separate rows
genre_ratings = genre_exploded.groupby('Genres')['imdbRating'].mean().reset_index()

# Print average ratings by genre
print("Average Ratings by Genre:")
print(genre_ratings)

# Step 4: Recommendation System
def recommend_movies(df, rating_threshold):
    return df[df['imdbRating'] > rating_threshold]

# Recommend movies with an IMDb rating greater than 8.0
recommendations = recommend_movies(df_movies, 8.0)
print("\nRecommended Movies:")
print(recommendations[['Title', 'imdbRating']])

# Step 5: Visualization
# Visualize average ratings by genre
plt.figure(figsize=(10, 6))
sns.barplot(x='Genres', y='imdbRating', data=genre_ratings)
plt.title("Average Movie Ratings by Genre")
plt.xticks(rotation=45)
plt.ylabel("Average IMDb Rating")
plt.xlabel("Genre")
plt.tight_layout()
plt.show()
```
